```python
from Bio.SeqIO.FastaIO import SimpleFastaParser
count = 0
total_len = 0
with open("BCAR1.fasta") as in_handle:
  for title, seq in SimpleFastaParser(in_handle):
    count += 1
    total_len += len(seq)

  print("%i records with total sequence length %i" % (count, total_len))
```
